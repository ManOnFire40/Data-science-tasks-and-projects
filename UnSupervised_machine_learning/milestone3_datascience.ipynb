```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('customers.csv')
```

```python
# Countplot of Gender

#making countplot
sns.countplot(data=data, x='Gender')
#naming title
plt.title('Gender Distribution')
#showing plot
plt.show()
#This plot will show the distribution of customers based on their gender. Insights gained: We can observe the relative proportions of male and female customers
```

```python
# Plotting the Age distribution

#making size
plt.figure(figsize=(10, 6))
#making histplot
sns.histplot(data['Age'], bins=20, kde=True)
#naming title
plt.title('Age Distribution')
#naming xlabel
plt.xlabel('Age')
#naming ylabel
plt.ylabel('Count')
#showing plot
plt.show()
#This plot will give us insights into the distribution of customers' ages. Insights gained: We can identify the age groups with the highest customer count and observe any significant patterns
```

```python
# Scatter plot of Income vs. Score

#splitting the data
sns.scatterplot(data=data, x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'])
#naming title
plt.title('Income vs. Score')
#naming xlabel
plt.xlabel('Income')
#naming ylabel
plt.ylabel('Score')
#showing plot
plt.show()
#This plot will help us understand the relationship between customer income and their score. Insights gained: We can look for any clusters or patterns in the scatter plot and determine if there are any correlations between income and score.

```

```python
from sklearn.cluster import KMeans

# Selecting features for clustering
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Creating the K-means model
kmeans = KMeans(n_clusters=5, random_state=42)

# Fitting the model to the data
kmeans.fit(features)

# Adding the cluster labels to the dataset
data['Cluster'] = kmeans.labels_

```

```python
# Scatter plot of Income vs. Score with cluster labels

#splitting the data
sns.scatterplot(data=data, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis')
#naming title
plt.title('Income vs. Score (Clustered)')
#naming xlabel
plt.xlabel('Income')
#naming ylabel
plt.ylabel('Score')
#showing plot
plt.show()
#This plot will display the clusters identified by the K-means algorithm, where each cluster is assigned a different color. Insights gained: We can observe the grouping of customers based on their income and score, as determined by the clustering algorithm
```
